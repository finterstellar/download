```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('09.2 KOSPI200 F 201712.xlsx', 'Sheet1', index_col='일자')
df.head()
```

```python
type(df.index)
```

```python
df.index = pd.to_datetime(df.index)
```

```python
type(df.index)
```

```python
def futures_price (S, r, d, T, t0):
    t = np.busday_count(t0, T) / 252
    F = S * np.exp((r-d)*t)
    return F
```

```python
T = pd.to_datetime('2017-12-14')     # 최종거래일
r = 0.0225     # 이자율
d = 0.015     # 배당률
for n in (df.index):
    t0 = n     
    S = df.loc[n, '지수']
    F = futures_price (S, r, d, T, t0)
    f = '{:,.2f}'.format(F)
    df.loc[n, '이론가1'] = f
```

```python
def S_futures_price (S, r, d, T, t0):
    t = (T - t0).days / 365
    F = S * (1 + (r-d) * t)
    return F
```

```python
T = pd.to_datetime('2017-12-14')     # 최종거래일
r = 0.0225     # 이자율
d = 0.015     # 배당률
for n in (df.index):
    t0 = n     
    S = df.loc[n, '지수']
    F = S_futures_price (S, r, d, T, t0)
    f = '{:,.2f}'.format(F)
    df.loc[n, '이론가2'] = f
```
